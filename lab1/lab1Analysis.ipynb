```python
%matplotlib notebook
import pandas as pd
```

```python
collatz = pd.read_csv('./part1.txt')
collatz.columns = ['Initial value (n)', 'Stopping Time']
ax1 = collatz.plot(x='Initial value (n)', y='Stopping Time', kind='scatter')
ax1.set_title("Collatz stopping time from initial value")
```

```python
ax = collatz.hist(column="Stopping Time", bins=50)
mean = collatz["Stopping Time"].mean()
ax[0, 0].set_title("Stopping Time for 1 <= n <= 1000")
ax[0, 0].set_xlabel("Number of iterations")
ax[0, 0].set_ylabel("Frequency")
ax[0, 0].text(140, 50, "Mean = {0:.2f}".format(mean))
```

```python
whileLoopIterations = pd.read_csv('./part2.txt', header=None).transpose()
whileLoopIterations.columns = ["Decrementing n", "Incrementing n", "Random n"]
ax3 = whileLoopIterations.plot()
ax3.set_xlabel("Solution attempt")
ax3.set_ylabel("Number of while loop iterations")
ax3.set_title("Collatz loop iterations with cache")
```

```python
unbounded = pd.read_csv('./part3a.txt', header=None)
unbounded.columns = ["n", "Stopping Time"]
ax_un = unbounded.hist(cumulative=True, column="n", histtype='stepfilled', bins=500)[0][0]
ax_un.set_xlabel("Max cache size")
ax_un.set_ylabel("Number of filled cache elements")
ax_un.set_title("")
```

```python
whileLoopIterationsUnbounded = pd.read_csv('./part3b.txt', header=None).transpose()
whileLoopIterationsUnbounded.columns = ["Decrementing n (unbound)", "Incrementing n (unbound)", "Random n (unbound)"]
```

```python
combined = pd.concat([whileLoopIterations, whileLoopIterationsUnbounded], axis=1)
ax5 = combined.plot()
ax5.set_xlabel("Solution attempt")
ax5.set_ylabel("Number of while loop iterations")
ax5.set_title("Collatz loop iterations with unbounded and bound cache")
```
