```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./greedy_results.csv")

plt.figure()
for m in [0, 1, 10, 300]:
    mdata = df.loc[df['m'] == m]
    plt.plot(mdata['n'], mdata['t'], '--', label="maxCommittedTrips = {}".format(m))
    
plt.xlabel("Number of drivers")
plt.ylabel("Total waiting time")
plt.legend()
```

```python
df = pd.read_csv("./greedy_results.csv")

plt.figure()
for m in [0, 1, 10, 300]:
    mdata = df.loc[df['m'] == m]
    mdata = mdata.loc[df['n'] <= 10]
    mdata = mdata.loc[1 < df['n']]
    plt.plot(mdata['n'], mdata['t'], 'o', alpha=0.3, label="maxCommittedTrips = {}".format(m))
    
plt.xlabel("Number of drivers")
plt.ylabel("Total waiting time")
plt.legend()
```

```python
df2 = pd.read_csv("./greedy_supplement.csv")
df1 = df.loc[df['m'] == 0]
plt.figure()
plt.plot(df2['n'], df2['t'], '-', alpha=0.8, label='original data set')
plt.plot(df1['n'], df1['t'], '-', alpha=0.8, label='supplementary data set')
plt.xlabel("Number of drivers")
plt.ylabel("Total waiting time")
plt.legend()
```

```python
df2 = pd.read_csv("./greedy_supplement.csv")
df2 = df2.loc[df2['n'] > 1]
df1 = df.loc[df['m'] == 0]
df1 = df1.loc[df['n'] > 1]

plt.figure()
plt.plot(df1['n'], df1['t'], 'o', alpha=0.5, label='original data set')
plt.plot(df2['n'], df2['t'], 'o', alpha=0.5, label='supplementary data set')
plt.xlabel("Number of drivers")
plt.ylabel("Total waiting time")
plt.legend()
```
